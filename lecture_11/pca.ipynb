```python
from sklearn.decomposition.pca import PCA
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
import numpy as np
```

```python
X, y = make_blobs(centers=2)
```

```python
plt.scatter(X[:, 0], X[:, 1])
plt.xlim(-20, 10)
plt.ylim(-20, 10)
```

```python
model = PCA()
```

```python
model.fit(X)
```

```python
model.components_
```

```python
X_mod = np.dot(X, model.components_.T)
```

```python
plt.scatter(X_mod[:, 0], X_mod[:, 1])

plt.xlim(-20, 10)
plt.ylim(-20, 10)
```

```python
X_gen = model.transform(X)
```

```python
plt.scatter(X_gen[:, 0], X_gen[:, 1])
plt.scatter(X_mod[:, 0], X_mod[:, 1])

plt.xlim(-20, 10)
plt.ylim(-20, 10)
```

```python
data = np.load("../datasets/mnist_train_small.npy")
```

```python
X, y = data[:, 1:], data[:, 0]
```

```python
model = PCA()
model.fit(X)
model.components_.shape
```

```python
plt.plot(model.explained_variance_)
```

```python
X_small = model.transform(X)
```

```python
plt.scatter(X_small[:, 0][y < 2], X_small[:, 1][y < 2], c= y[y < 2])
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(
...     X_small, y, test_size=0.33, random_state=42)
```

```python
classifier = DecisionTreeClassifier(max_depth=10)

classifier.fit(X_train, y_train)

classifier.score(X_test, y_test)

classifier.score(X_train, y_train)
```

```python
classifier = RandomForestClassifier(max_depth=10)

classifier.fit(X_train, y_train)

print(classifier.score(X_test, y_test), classifier.score(X_train, y_train))

```
