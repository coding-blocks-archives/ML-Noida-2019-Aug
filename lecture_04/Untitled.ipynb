```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = np.zeros([300, 300, 3])
```

```python
plt.imshow(img)
```

```python
img[:100] = np.array([1, 0, 0])
img[100:200] = np.array([0, 1, 0])
img[200:] = np.array([0, 0, 1])

img[50:-50, 50:-50] = np.array([.5, .5, .5])
```

```python
plt.imshow(img)
```
