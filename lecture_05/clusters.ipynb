```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
mean1 = [4, 4]
covar1 = [[1, -1], [-1, 5]]

mean2 = [-1, -1]
covar2 = [[1, 0], [0, 1]]
```

```python
d1 = np.random.multivariate_normal(mean1, covar1, 1000)
d2 = np.random.multivariate_normal(mean2, covar2, 1000)
```

```python
d1.shape
```

```python
plt.scatter(d1[:, 0], d1[:, 1])
```

```python
d1[:, 0].shape
```

```python
plt.figure()
plt.scatter(d1[:, 0], d1[:, 1], label="cluster-1")
plt.scatter(d2[:, 0], d2[:, 1], label="cluster-2")
plt.xlim(-10, 10)
plt.ylim(-10, 10)
plt.xlabel("longitude")
plt.ylabel("latitube")
plt.title("Poore 20 number")
plt.legend()
plt.show()
```

```python
np.var(d1[:, 0])
```

```python
d1.shape
```

```python
d2.shape
```

```python
X = np.vstack([d1, d2])

y = np.ones([len(d1) + len(d2)], dtype=int)
y[len(d2):] = 2
```

```python
y
```

```python
plt.figure()
plt.scatter(X[:, 0], X[:, 1], c=y)
plt.xlim(-10, 10)
plt.ylim(-10, 10)
plt.xlabel("longitude")
plt.ylabel("latitube")
plt.title("Poore 20 number")
plt.legend()
plt.show()
```
