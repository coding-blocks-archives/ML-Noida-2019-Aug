```python
import matplotlib.pyplot as plt
import numpy as np
from keras.preprocessing.image import load_img
```

```python
img = load_img("../datasets/nums/one.jpeg", target_size=(300, 400), grayscale=True)
```

```python
np_img = np.array(img)
```

```python
np_img.dtype
```

```python
def apply_kernel(img, kernel):
    
#     target_img = np.zeros([img.shape[0] - kernel.shape[0] + 1, img.shape[1] - kernel.shape[1] + 1, img.shape[2]], dtype=np.uint8)
    
    target_img = np.zeros([img.shape[0] - 2, img.shape[1] - 2], dtype=np.uint8)
    
    for row in range(target_img.shape[0]):
        for col in range(target_img.shape[1]):
            cut = img[row:row+3, col:col+3]
            target_img[row, col] = np.abs((cut * kernel).mean())
    
    return target_img
            
```

```python
target = apply_kernel(np_img, np.array([[-1, -1, -1], [0, 0, 0], [1, 1, 1]]))
```

```python
plt.imshow(target, cmap="gray")
```

```python
target = apply_kernel(np_img, np.array([[-1, 0, 1], [-1, 0, 1], [-1, 0, 1]]))
```

```python
plt.imshow(target, cmap="gray")
```
