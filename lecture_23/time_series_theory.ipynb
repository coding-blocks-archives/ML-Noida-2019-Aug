```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
```

```python
apple_df = pd.read_csv("../datasets/AAPL.csv")
```

```python
apple_df.head()
```

```python
apple_high_unproc = apple_df.iloc[:, 2:3].values
```

```python
plt.plot(apple_high_unproc)
```

```python
print("max", apple_high_unproc.max())
print("min", apple_high_unproc.min())
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
apple_high_proc = scaler.fit_transform(apple_high_unproc)
```

```python
plt.plot(apple_high_proc)
```

```python
print("max", apple_high_proc.max())
print("min", apple_high_proc.min())
```

```python
moving_avg = []
for i in range(0, len(apple_high_proc)):
    moving_avg.append(apple_high_proc[i:i+20].mean())
```

```python
plt.plot(moving_avg)
```

```python
high_without_trend = apple_high_proc - np.array(moving_avg).reshape(-1, 1)
```

```python
plt.plot(high_without_trend)
```

```python
level_1 = high_without_trend[2:] - high_without_trend[0:-2]
```

```python
plt.plot(level_1)
```

```python
level_2 = level_1[2:] - level_1[:-2]
```

```python
plt.plot(level_2)
```
